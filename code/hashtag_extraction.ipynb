```python
#Importing libraries
import pandas as pd 
from selenium.webdriver.common.keys import Keys
from selenium import webdriver
import time
import numpy as np
import csv
from selenium import webdriver
```

```python
driver = webdriver.Chrome(executable_path=r'C:\Program Files (x86)\Google\Chrome\Application\81.0.4044.129\default_apps\chromedriver.exe')
data = pd.DataFrame(columns = ['trace_title','genere','hashtag'])
with open(r"C:\Users\clara\Desktop\df_tot.csv",encoding = 'utf8') as url_obj:
    reader = csv.DictReader(url_obj,delimiter = ',')
    for line in reader:
        trace_title=line["trace_title"]
        genere=line["genre"]
        url=line["link"]
        print (url)
        driver.get(url)
        df = pd.DataFrame(columns = ['trace_title', 'genre', 'link'])
        elem = driver.find_element_by_tag_name("body")
        prendi = driver.find_elements_by_class_name("sc-tag-group")
        hashtags = []
        for c in prendi:
            link = c.find_elements_by_tag_name('a')
            for i in link:
                ancora = i.find_element_by_class_name('sc-truncate')
                hashtags.append(ancora.get_attribute('innerHTML'))
            data = data.append({'trace_title':trace_title,'genere':genere,'hashtag':hashtags},ignore_index=True)
        time.sleep(3)
driver.quit()     
```
